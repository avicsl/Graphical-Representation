# **A. Demography**

***1. Population***

```python
import matplotlib.pyplot as plt

def plot_bar_pie(title, categories, values, bar_colors, pie_colors):
    x_pos = range(len(categories))
    bar_width = 0.6

    # Font settings
    title_font = {'fontsize': 12, 'fontweight': 'bold', 'color': '#FF69B4'}
    label_font = {'fontsize': 9, 'color': '#8A2BE2'}

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Bar chart
    bars = ax1.bar(x_pos, values, width=bar_width, color=bar_colors, edgecolor='white', linewidth=1.5)
    ax1.set_title(title, **title_font)
    ax1.set_ylabel("Population Magnitude", **label_font)
    ax1.set_ylim(0, max(values) + 5)
    ax1.set_xticks(x_pos)
    ax1.set_xticklabels(categories, rotation=0, ha='center', color='#FF69B4')
    ax1.tick_params(axis='y', labelcolor='#8A2BE2')
    ax1.grid(axis='y', linestyle='--', alpha=0.5)

    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', color='#8A2BE2', fontsize=8)

    # Pie chart
    total_population = 91
    percentages = [(v / total_population) * 100 for v in values]
    ax2.pie(values, labels=[f"{cat}\n{perc:.1f}%" for cat, perc in zip(categories, percentages)],
            startangle=140, colors=pie_colors, textprops={'color': '#8A2BE2'})
    ax2.set_title("Percentage Distribution", **title_font)
    ax2.axis('equal')

    plt.tight_layout()
    plt.show()

# DEMOGRAPHY DATA (Pastel Palette)
categories = ["Children 0-5 \nyears old", "Children 6-12 \nyears old", "Children 13-16 \nyears old", "Children 17+ \nyears old"]
# Known values
value_0_5 = 8
value_6_12 = 19
value_13_16 = 11
# Calculate 17+ years old
total_population = 91
value_17_plus = total_population - (value_0_5 + value_6_12 + value_13_16)

values = [value_0_5, value_6_12, value_13_16, value_17_plus]
# pastel colors
bar_colors = ["#FDEF81", "#FBCFE0", "#A5F1E9", "#D8A7D8"]
pie_colors = ["#FDEF81", "#FBCFE0", "#A5F1E9", "#D8A7D8"]

plot_bar_pie("Population \n Total Population = 91", categories, values, bar_colors, pie_colors)
```

***7. Income and Livelihood***

```python
import matplotlib.pyplot as plt

def plot_bar_pie(title, categories, values, bar_colors, pie_colors):
    x_pos = range(len(categories))
    bar_width = 0.6

    # Font settings (pastel style, smaller for readability)
    title_font = {'fontsize': 11, 'fontweight': 'bold', 'color': '#FF69B4'}  # pastel pink
    label_font = {'fontsize': 8, 'color': '#8A2BE2'}  # pastel violet

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Bar chart
    bars = ax1.bar(x_pos, values, width=bar_width, color=bar_colors, edgecolor='white', linewidth=1.5)
    ax1.set_title(title, **title_font)
    ax1.set_ylabel("Population Magnitude", **label_font)
    ax1.set_ylim(0, max(values) + 5)
    ax1.set_xticks(x_pos)
    ax1.set_xticklabels(categories, rotation=0, ha='center', color='#FF69B4', fontsize=7)
    ax1.tick_params(axis='y', labelcolor='#8A2BE2', labelsize=7)
    ax1.grid(axis='y', linestyle='--', alpha=0.5)

    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval),
                 ha='center', va='bottom', color='#8A2BE2', fontsize=7)

    # Pie chart
    total_population = sum(values)
    percentages = [(v / 22) * 100 for v in values]  # Calculate percentage by dividing by 22
    wedges, texts = ax2.pie(
        values,
        labels=[f"{cat}\n{perc:.1f}%" for cat, perc in zip(categories, percentages)],
        startangle=140,
        colors=pie_colors,
        textprops={'color': '#8A2BE2', 'fontsize': 7}
    )
    ax2.set_title("Percentage Distribution", **title_font)
    ax2.axis('equal')

    # Move the pie chart a little lower
    box = ax2.get_position()
    ax2.set_position([box.x0, box.y0 - 0., box.width, box.height])

    plt.tight_layout()
    plt.show()

# INCOME AND LIVELIHOOD DATA (Pastel Palette)
categories = [
    "Households with income \nbelow poverty threshold",
    "Households with income \nbelow food threshold",
    "Households who \nexperienced food shortage",
    "Unemployed members \nof the labor force"
]
values = [15, 10, 2, 5]

# PASTEL colors (same style as demography)
bar_colors = ["#FDEF81", "#FBCFE0", "#A5F1E9", "#D8A7D8"]
pie_colors = bar_colors

plot_bar_pie("Income and Livelihood\nTotal Household = 22", categories, values, bar_colors, pie_colors)
```
